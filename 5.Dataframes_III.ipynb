# .83 A Review of .apply on single column

```python
import pandas as pd
bond = pd.read_csv('./csv/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def convert_to_str_and_add_mil(number):
    return str(number)+" "+"Millions"

bond["Box Office"]=bond["Box Office"].apply(convert_to_str_and_add_mil)
```

```python
bond['Budget']=bond['Budget'].apply(convert_to_str_and_add_mil)
```

```python
bond.head()
```

```python
columns = ["Box Office", "Bond Actor Salary","Budget"]

for col in columns:
    bond[col]=bond[col].apply(convert_to_str_and_add_mil)
```

```python
bond.head(3)
```

# .84 The .apply() method with Row values

```python
import pandas as pd
bond = pd.read_csv('./csv/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    if actor == 'Pierce Brosnan':
        return 'The Best'
    if actor == 'Roger Moore' and budget > 40:
        return 'Enjoyable'
    else:
        return 'No clue'

bond.apply(good_movie, axis = 'columns')   # since we are moving across column we should use columns and not rows  
```

# .85 the .copy method

```python
import pandas as pd
bond = pd.read_csv('./csv/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```
